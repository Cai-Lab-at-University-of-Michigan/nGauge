```python
import matplotlib
from matplotlib import pyplot as plt

from glob import glob
import seaborn as sns
import matplotlib.pyplot as plt

#%matplotlib inline

from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Arial']})
```

```python
import ngauge
from ngauge import Neuron
```

```python
# Open a collection of files
files = [ Neuron.from_swc(x) for x in glob('ye_swc/*.swc') ]
```

```python
# Do the plot
sns.boxplot( [ x.max_branching_order() for x in files ], color=(52/256, 66/256, 123/256) )

# Reformat so it is the right size
fig = plt.gcf()
fig.set_size_inches(2.4,0.8)
fig.subplots_adjust(bottom=.3,left=.15)
plt.savefig('3E.pdf', dpi=300, transparent=True)
```

```python
# Do the plot
sns.boxplot( [ x.soma_volume() for x in files ], color=(52/256, 66/256, 123/256) )

# Reformat so it is the right size
fig = plt.gcf()
fig.set_size_inches(2.4,0.8)
fig.subplots_adjust(bottom=.3,left=.15)
plt.savefig('3F.pdf', dpi=300, transparent=True)
```

```python
# Select a neuron
file = files[7]

# Plot 
fig = file.plot(fig=None, ax=None, color=(52/256, 66/256, 123/256) )

# Format and save
ax = fig.get_axes()[0]
ax.axis('off')
plt.savefig( 'neuron_plot.pdf', transparent=True )
```
